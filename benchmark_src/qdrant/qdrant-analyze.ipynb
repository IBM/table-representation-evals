```python
import pandas as pd
from qdrant_client import QdrantClient
from pathlib import Path
```

```python
path = Path.cwd().parent.parent / "cache" / "qdrant_storage"
client = QdrantClient(path=str(path))
client.get_collections()
```

```python
collection_name = client.get_collections().collections[2].name

collection_name
```

```python
collection = client.get_collection(collection_name)
client.scroll(collection_name=collection_name, limit=5,with_vectors=True)
```

```python
per_database_count = {}
print(client.scroll(collection_name=collection_name, limit=2)[1])
for point in client.scroll(collection_name=collection_name, limit=10000)[0]:
    db_id = point.payload['database_id']
    if db_id not in per_database_count:
        per_database_count[db_id] = 0
    per_database_count[db_id] += 1

df = pd.DataFrame.from_dict(per_database_count, orient='index', columns=['count']).sort_values(by='count', ascending=False)
df
```
